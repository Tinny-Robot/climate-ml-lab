## 3. Engineered Features <a id="engineered-features"></a>

### 3.1 Lag Features

**Purpose:** Capture temporal dependencies by including previous day's values

**Methodology:** For each numeric variable `X`, create `X_lag1 = X.shift(1)`

**Rationale:** Weather exhibits strong autocorrelation - today's weather strongly influences tomorrow's weather

```python
# Lag features dictionary
lag_features = pd.DataFrame([
    {
        'Feature Name': 'temperature_c_lag1',
        'Type': 'Float',
        'Unit': '°C',
        'Description': 'Temperature from previous day (t-1)',
        'Formula': 'temperature_c.shift(1)',
        'Importance': 'Very High - Strongest predictor for next-day temperature',
        'Correlation with Target': '~0.95+'
    },
    {
        'Feature Name': 'humidity_pct_lag1',
        'Type': 'Float',
        'Unit': '%',
        'Description': 'Humidity from previous day (t-1)',
        'Formula': 'humidity_pct.shift(1)',
        'Importance': 'High - Indicates atmospheric moisture conditions',
        'Correlation with Target': 'Moderate negative'
    },
    {
        'Feature Name': 'wind_speed_mps_lag1',
        'Type': 'Float',
        'Unit': 'm/s',
        'Description': 'Wind speed from previous day (t-1)',
        'Formula': 'wind_speed_mps.shift(1)',
        'Importance': 'Moderate - Affects heat transfer',
        'Correlation with Target': 'Weak'
    },
    {
        'Feature Name': 'precipitation_mm_lag1',
        'Type': 'Float',
        'Unit': 'mm',
        'Description': 'Precipitation from previous day (t-1)',
        'Formula': 'precipitation_mm.shift(1)',
        'Importance': 'Moderate - Indicates recent rainfall',
        'Correlation with Target': 'Weak to moderate'
    },
    {
        'Feature Name': 'surface_pressure_hpa_lag1',
        'Type': 'Float',
        'Unit': 'hPa',
        'Description': 'Surface pressure from previous day (t-1)',
        'Formula': 'surface_pressure_hpa.shift(1)',
        'Importance': 'High - Strong influence on temperature',
        'Correlation with Target': 'Moderate positive'
    }
])

display(lag_features.style.set_properties(**{'text-align': 'left'}).set_table_styles(
    [{'selector': 'th', 'props': [('text-align', 'left'), ('background-color', '#A23B72'), ('color', 'white')]}]
))
```

## 4. Target Variable <a id="target-variable"></a>

### 4.1 Target Definition

**Name:** `target_temp_next_day`

**Formula:** `temperature_c.shift(-1)`

**Description:** The temperature for the following day (next-day temperature forecasting)

**Type:** Continuous (Float)

**Unit:** °C (Celsius)

**Forecasting Horizon:** 1 day (24 hours ahead)

**Task Type:** Regression (predict continuous value)

### 4.2 Target Statistics

```python
# Load full processed data for statistics
try:
    full_df = pd.read_csv('../data/processed/clean_weather.csv', parse_dates=['date'], index_col='date')
    target = full_df['temperature_c'].shift(-1).dropna()
    
    target_stats = pd.DataFrame({
        'Statistic': ['Count', 'Mean', 'Std Dev', 'Min', '25th Percentile', 'Median (50th)', '75th Percentile', 'Max', 'Range'],
        'Value': [
            f"{len(target):,}",
            f"{target.mean():.2f}°C",
            f"{target.std():.2f}°C",
            f"{target.min():.2f}°C",
            f"{target.quantile(0.25):.2f}°C",
            f"{target.median():.2f}°C",
            f"{target.quantile(0.75):.2f}°C",
            f"{target.max():.2f}°C",
            f"{target.max() - target.min():.2f}°C"
        ]
    })
    
    display(target_stats.style.hide(axis='index').set_table_styles(
        [{'selector': 'th', 'props': [('text-align', 'left'), ('background-color', '#06A77D'), ('color', 'white')]}]
    ))
    
except FileNotFoundError:
    print("⚠ Data not available for target statistics")
```

### 5.1 Evaluation Metrics Definitions

## 6. Data Quality Indicators <a id="data-quality"></a>

### 6.1 Data Completeness

```python
# Data quality assessment
try:
    full_df = pd.read_csv('../data/processed/clean_weather.csv', parse_dates=['date'], index_col='date')
    
    # Calculate completeness
    total_rows = len(full_df)
    expected_days = (full_df.index.max() - full_df.index.min()).days + 1
    completeness = (total_rows / expected_days) * 100
    
    quality_report = pd.DataFrame([
        {'Indicator': 'Total Records', 'Value': f"{total_rows:,}"},
        {'Indicator': 'Date Range', 'Value': f"{full_df.index.min().date()} to {full_df.index.max().date()}"},
        {'Indicator': 'Expected Days', 'Value': f"{expected_days:,}"},
        {'Indicator': 'Completeness', 'Value': f"{completeness:.1f}%"},
        {'Indicator': 'Missing Values (Post-Processing)', 'Value': f"{full_df.isnull().sum().sum()}"},
        {'Indicator': 'Duplicate Dates', 'Value': f"{full_df.index.duplicated().sum()}"},
        {'Indicator': 'Data Quality Status', 'Value': '✓ High' if completeness > 99 and full_df.isnull().sum().sum() == 0 else '⚠ Review'}
    ])
    
    display(quality_report.style.hide(axis='index').set_table_styles(
        [{'selector': 'th', 'props': [('text-align', 'left'), ('background-color', '#6A994E'), ('color', 'white')]}]
    ))
    
except FileNotFoundError:
    print("⚠ Data not available for quality assessment")
```

- Report data collection dates
- Document any preprocessing steps
